```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000

a = 3

x = np.random.power(a,size=n)

plt.hist(x,bins=100,alpha=0.5,edgecolor="black")
plt.ylabel("Amount per bin")
plt.xlabel("x")
plt.savefig("histogram_power_function.pdf",bbox_inches="tight",facecolor="white")
```
